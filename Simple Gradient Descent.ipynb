```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    
    return x**4 - 4*x**2 +5
```

```python
def df(x):
    
    return 4*x**3 - 8*x
```

```python
x_1 = np.linspace(start = -2, stop = 2, num = 300 )
plt.figure(figsize = (15, 7))


plt.subplot(1,2,1)

plt.plot(x_1, f(x_1), c = 'navy', linewidth = 2)
#plt.scatter(x_1, f(x_1), c='red')
plt.grid()


plt.subplot(1,2,2)

plt.plot(x_1, df(x_1), c = 'red', linewidth = 2)
plt.grid()


plt.show()
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier, precision , max_iter):
    
    new_x = initial_guess
    
    x_list = [new_x]
    
    slope_list = [derivative_func(new_x)]
    
    
    for i in range(max_iter):
        
        
        previous_x = new_x
        
        gradient = derivative_func(previous_x)
        
        new_x = previous_x - multiplier * gradient
        
        step_size = abs(new_x - previous_x)
        
        x_list.append(new_x)
        
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            
            break 
            
    return new_x, x_list, slope_list
    
```

```python
local_min, list_x , deriv_list =gradient_descent(derivative_func = df,initial_guess =2, multiplier = 0.1, precision=0.00001 , max_iter = 1000)
```
